#### 函数库导入

```python
import torch
import torch.nn as nn
from basic_model import DeepModel_single, gradients
```

#### 模型参数
$$
    y = f(x;\xi)\\
    x: N_1, N_2, N_3, ..., M 矩阵\\
    y: N_1, N_2, N_3, ..., N 矩阵\\
$$


```python
# 构建网络
Net = DeepModel_single([2] + [32] *4 + [5], active=nn.ReLU())
print(Net.layers)
# 给定输入
x1 = torch.randn((100, 2), dtype=torch.float32)
# 计算输出
y1 = Net(x1)
print(y1.shape)
print(y1)
```

```python
# 给定输入
x2 = torch.randn((100, 300, 2), dtype=torch.float32)
# 计算输出
y2 = Net(x2)
print(y2.shape)
print(y2)
```

```python
# 给定输入
x3 = torch.randn((100, 300, 5, 2), dtype=torch.float32)
# 计算输出
y3 = Net(x3)
print(y3.shape)
print(y3)
```

### 一阶导数

```python
# 给定输入
x = torch.ones((10, 5, 4, 3, 2), dtype=torch.float32)
# 注意必须将自变量的梯度置为true，才能求自动微分
x.requires_grad_(True)
# 计算输出，前向过程
y = Net(x)
# 计算y对x的一阶微分，反向过程
dydx = gradients(y, x)
print(dydx.shape)
print(dydx)
```

### 二阶导数

```python
# 计算y对x的二阶微分，反向过程，注意求导机制，保证除最后一个维度外对x都是独立的
d2ydx2 = gradients(dydx[..., (0,)], x)
print(d2ydx2.shape)
print(d2ydx2)
```
